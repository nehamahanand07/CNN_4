```python
#we are doing supervised ML classification, we need input and output
flat_data_arr=[] #collect input
target_arr=[] #output
categories=['with_mask','without_mask']
datadir='/content/drive//MyDrive'
```

```python
#LOADING CATEGORIES

for i in categories:
  print('Loading categories...')
  path=os.path.join(datadir,i)
  for img in os.listdir(path):
    img_arr=imread(os.path.join(path,img))
    img_resize=resize(img_arr,(150,150,3))

    img_gray=rgb2gray(img_resize)
    img_gray=img_gray.reshape(150,150,1)


    flat_data_arr.append(img_gray)
    target_arr.append(categories.index(i))
  print('Loaded completely',i)
```

```python
#converting input and output as arrays
flat_data=np.array(flat_data_arr)
target_data=np.array(target_arr)
```

```python
# #model creation sequential

# from keras.layers import Conv2D, MaxPool2D, Flatten, Dense

# model=Sequential()

# model.add(Conv2D(32,(3,3),activation='relu',input_shape=(150,150,1)))

# model.add(MaxPool2D(2,2))

# model.add(Flatten())

# model.add(Dense(64,activation='relu'))

# model.add(Dense(2,activation='softmax'))

#modelcreation
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, Flatten
model = Sequential()
model.add(Conv2D(64, (3, 3), activation='relu', input_shape=(150, 150, 1)))  # Increased filters
model.add(MaxPool2D(2, 2))
model.add(Conv2D(128, (3, 3), activation='relu'))  # Added another convolutional layer
model.add(MaxPool2D(2, 2))
model.add(Flatten())
model.add(Dense(256, activation='relu'))  # Increased neurons
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=15,batch_size=256)
```

```python
path1='/content/wm_1.jpg'
img=imread(path1)
img=resize(img,(150,150,3))
img_gray=rgb2gray(img)
img_gray=img_gray.reshape(150,150,1)
prediction = model.predict(np.expand_dims(img_gray, axis=0)) # Adding batch dimension
print(f"Raw prediction output: {prediction}")
threshold = 0.5
predicted_class = "mask" if prediction[0][0] >= threshold else "no_mask"
print(f"Predicted class: {predicted_class}")
```
